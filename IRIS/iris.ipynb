```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
tf.reset_default_graph()
```

```python
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',sep=',',names=['sepal_length','sepal_width','petal_length','petal_width','iris_class'])
```

```python
data
```

```python
data=data.sample(frac=1).reset_index(drop=True)
#print(data)
all_x=data[['sepal_length','sepal_width','petal_length','petal_width']]
#print(all_x)
min_max_scaler=preprocessing.MinMaxScaler()
all_x=min_max_scaler.fit_transform(all_x)
all_y=pd.get_dummies(data.iris_class)# one hot encode
train_x,test_x,train_y,test_y=train_test_split(all_x,all_y,test_size=1/3)
nx=np.shape(train_x)[1]
ny=np.shape(train_y)[1]# the labels are one hot encoded

```

```python
print(train_y[0:4])
```

```python
tf.reset_default_graph()
learning_rate=0.01
h1=10
h2=20
h3=10
g=tf.Graph()
with g.as_default():
    x=tf.placeholder(tf.float32,[None,nx],name='x')
    y=tf.placeholder(tf.float32,[None,ny],name='y')
    fully_connected1=tf.contrib.layers.fully_connected(inputs=x,num_outputs=h1,activation_fn=tf.nn.relu,scope='Fully_Conn1')
    fully_connected2=tf.contrib.layers.fully_connected(inputs=fully_connected1,num_outputs=h2,activation_fn=tf.nn.relu,scope='Fully_Conn2')
    fully_connected3=tf.contrib.layers.fully_connected(inputs=fully_connected2,num_outputs=h3,activation_fn=tf.nn.relu,scope='Fully_Conn3')
    prediction_layer=tf.contrib.layers.fully_connected(inputs=fully_connected3,num_outputs=ny,activation_fn=tf.nn.softmax,scope='Output_layer')
    cost=tf.losses.softmax_cross_entropy(onehot_labels=y,logits=prediction_layer,scope='cost_function')
    correct_prediction=tf.equal(tf.argmax(prediction_layer,1,name='argmax_pred'),tf.argmax(y,1,name='Y_pred'),name='correct_pred')
    accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32,name='Cast_corr_pred'),name='accuracy')
    optimizer=tf.train.AdagradOptimizer(learning_rate,name="Optimizer").minimize(cost)
    
```

```python
with tf.Session(graph=g) as sess:
    init=tf.global_variables_initializer()
    sess.run(init)
    training_epochs=3000
    for epoch in range(training_epochs):
        _,c=sess.run([optimizer,cost],feed_dict={x:train_x,y:train_y})
    #to be used outside the training loop########################
    pred_bools=correct_prediction.eval({x:test_x,y:test_y})
    ###########only want the final val #####################
    print('accuracy {}'.format(accuracy.eval({x:test_x,y:test_y})))
    print('loss {}'.format(c))
    #######################################################################
    #### will work when written inside the session graph
    layers={v.op.name: v for v in tf.trainable_variables()}
    print(layers)
    
        
    
```

## SIMPLE  Residual net
### for better accuaracy on same data

```python
tf.reset_default_graph()
learning_rate=0.01
h1=10
h2=10
h3=10
g1=tf.Graph()
with g1.as_default():
    x=tf.placeholder(tf.float32,[None,nx],name='r_input')
    y=tf.placeholder(tf.float32,[None,ny],name='r_output')
    hiddenlayer1=tf.contrib.layers.fully_connected(inputs=x,num_outputs=h1,activation_fn=tf.nn.relu)

    hiddenlayer2=tf.contrib.layers.fully_connected(inputs=hiddenlayer1,num_outputs=h2,activation_fn=tf.nn.relu)
    hiddenlayer3=tf.contrib.layers.fully_connected(inputs=hiddenlayer2,num_outputs=h3,activation_fn=None)
    hiddenlayer3=tf.nn.relu(hiddenlayer3+hiddenlayer1)
    predictionlayer=tf.contrib.layers.fully_connected(hiddenlayer3,num_outputs=ny,activation_fn=tf.nn.softmax)
    cost=tf.losses.softmax_cross_entropy(onehot_labels=y,logits=predictionlayer)
    optimizer=tf.train.AdamOptimizer(learning_rate).minimize(cost)
    correct_prediction=tf.equal(tf.argmax(predictionlayer,axis=1,name='argmax_pred'),tf.argmax(y,axis=1,name='actual_y'))
    accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32,name='Cast_corr_pred'),name='accuracy')
```
